```python
titanic ['Age']=titanic['Age'].interpolate()
```

```python
titanic.info()
```

```python
#To simplify the process, It's wise to clean the data. For this process, I'm dropping Name, SibSp, Parch, Ticket and Embarked. 
#I'm dropping names, SibSp, Parch and Ticket as they are distinct categories and won't be needed for my analysis as of now. 
#Clean data would help to streamline the analysis and focus on specific questions that i'm looking in this datafile.

titanic=titanic.drop(columns='Name')
titanic=titanic.drop(columns='SibSp')
titanic=titanic.drop(columns='Parch')
titanic=titanic.drop(columns='Ticket')
titanic=titanic.drop(columns='Embarked')
titanic=titanic.drop(columns='Cabin')
titanic.head()
```

```python
#Analysis Questions

#1. Are Upper class passengers more liekly to survive the titanic? (varibales = Pclass and Survived)
#2. Are people paying higer ticket price more likely to survive the titanic? (variables = Fare and Survived)
#3. Are Women more likely to survive?  (Variables = Gender and Survived)
#4. Passengers belonging to which age group are more likely to survive the titanic? (variables = Age, Sex and Survived)
#5. Are childen more likely to survive the titanic? (Varibles = Age and Survival)
```

```python
#Broadly looking at the figures 

sns.countplot(titanic.Sex).set_title('Distribution of sexes')
plt.show()
sns.countplot(titanic.Pclass).set_title('Distribution of classes')
plt.show()
sns.distplot (titanic.Age, bins = 50).set_title('Distribution of Age')
plt.show()
sns.distplot (titanic.Fare, bins = 50).set_title('Distribution of Fare')
plt.show()
```

```python
#Observations from the above graphs 

#1. Gender - Number of males travelling are more than females 
#2. Class - PAssengers are more likely to travel in the third class followed by first class. 
#3. Age - Number of elederly people travelling is very less. People in the age group of 20-30 are travelling more.
#4. Fares - Majority of the passengers payed less fare as compared to others 
```

```python
#Relationship between Class and Survival

fig, ax = plt.subplots(figsize=[4,2])
g = sns.countplot(x='Pclass', hue='Survived', data=titanic)
g.set_title('Distribution of Survival by Ticket Class', fontsize=9.5)
g.set_xlabel('Ticket Class')
g.set_ylabel('Frequency')
plt.show()

# Survived = 0 -No, 1 -Yes

#It can be seen that passengers travelling in the first class are more likley to survive followed by passengers travelling in the second class. 
#The crosstab below reflects the death of passengers more in the third class (372 died).
```

```python
pd.crosstab(titanic.Survived, titanic.Pclass)
```

```python
titanic.Pclass.corr(titanic.Survived)

#It can be noticed that the class correlates strongly with the survival.
```

```python
#Relationship between Fare and Survival

#It is already known that the passengers travelling in the first class are more likely to survive as compare to the passengers travelling in the second and third class. 
#It is evident from the graph below that the frequency of the passengers are more likely to survive who have paid higher fares. 

g = sns.FacetGrid(titanic, hue='Survived', height=2.6, aspect=2.1, sharex=False)
g.map(plt.hist, 'Fare', alpha=.6)
g.fig.suptitle('Distribution of Survival by Fare', fontsize=9.5, y=1.01)
g.set_ylabels('Frequency')
plt.show()
```

```python
titanic.Fare.corr(titanic.Survived)

#The fare correlates positively with the survival rate. The correlation is not as strong as Pclass but cannot be dismissed for the analysis
```

```python
#Relationship between Gender and Survival 

#From the above analysis we already know that there are more male passengers as compared to female passengers.
#From the below graph we can see that females are more likely to survive the titanic as compared to males. One of the prime reason of this could be that the passengers travelling in the titanic are more males than females. 
#From the total 891 passengers travelling, 468 male passengers died as comapred to only 81 female passengers.


fig, ax = plt.subplots(figsize=[4,2])
g = sns.countplot(x='Sex', hue='Survived', data=titanic)
g.set_title('Distribution of Survival by Sex', fontsize=9.5)
g.set_xlabel('Sex')
g.set_ylabel('Frequency')
plt.show()
```

```python
pd.crosstab(titanic.Sex, titanic.Survived)
```

```python
#Relationship between Children and Survival

#In the above graph we have already seen that the survival rate was lowest for passengers in the age group of 20-30. 
#The survival rate for children is much more than adults.

fig, ax = plt.subplots(figsize=[4,2])
ax.hist(titanic.Age[titanic.Survived == 0].dropna(), alpha=.6)
ax.hist(titanic.Age[titanic.Survived == 1].dropna(), alpha=.6)
ax.set_title('Distribution of Survival by Age', fontsize=9)
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
ax.legend(['0','1']).set_title('Survived', prop={'size':'small'})
plt.show()
```

```python
titanic.Age.corr(titanic.Survived)

#There is a very strong correlation betwen age and survival. 
```
